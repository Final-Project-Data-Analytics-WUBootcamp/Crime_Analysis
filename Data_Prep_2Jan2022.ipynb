```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
#Read in csvs, concatenate each years' unique dataframe together to consolidate data

offender_2020 = pd.read_csv("2020_NIBRS_OFFENDER.csv")
offender_2019 = pd.read_csv("2019_NIBRS_OFFENDER.csv")
offender_2018 = pd.read_csv("2018_NIBRS_OFFENDER.csv")
offender_2017 = pd.read_csv("2017_NIBRS_OFFENDER.csv")
offender_frames = [offender_2020,offender_2019,offender_2018,offender_2017]
offender_all = pd.concat(offender_frames)
#offender_all.head()

arrestee_2020 = pd.read_csv("2020_NIBRS_ARRESTEE.csv")
arrestee_2019 = pd.read_csv("2019_NIBRS_ARRESTEE.csv")
arrestee_2018 = pd.read_csv("2018_NIBRS_ARRESTEE.csv")
arrestee_2017 = pd.read_csv("2017_NIBRS_ARRESTEE.csv")
arrestee_frames = [arrestee_2020,arrestee_2019,arrestee_2018,arrestee_2017]
arrestee_all = pd.concat(arrestee_frames)
#arrestee_all.head()

incident_2020 = pd.read_csv("2020_NIBRS_incident.csv")
incident_2019 = pd.read_csv("2019_NIBRS_incident.csv")
incident_2018 = pd.read_csv("2018_NIBRS_incident.csv")
incident_2017 = pd.read_csv("2017_NIBRS_incident.csv")
incident_frames = [incident_2020,incident_2019,incident_2018,incident_2017]
incident_all = pd.concat(incident_frames)
#incident_all.head()

offense_2020 = pd.read_csv("2020_NIBRS_OFFENSE.csv")
offense_2019 = pd.read_csv("2019_NIBRS_OFFENSE.csv")
offense_2018 = pd.read_csv("2018_NIBRS_OFFENSE.csv")
offense_2017 = pd.read_csv("2017_NIBRS_OFFENSE.csv")
offense_frames = [offense_2020,offense_2019,offense_2018,offense_2017]
offense_all = pd.concat(offense_frames)
#offense_all.head()

victim_2020 = pd.read_csv("2020_NIBRS_VICTIM.csv")
victim_2019 = pd.read_csv("2019_NIBRS_VICTIM.csv")
victim_2018 = pd.read_csv("2018_NIBRS_VICTIM.csv")
victim_2017 = pd.read_csv("2017_NIBRS_VICTIM.csv")
victim_frames = [victim_2020,victim_2019,victim_2018,victim_2017]
victim_all = pd.concat(victim_frames)
#victim_all.head()

agency_2020 = pd.read_csv("2020_agencies.csv")
agency_all = agency_2020[['AGENCY_ID','PUB_AGENCY_NAME','AGENCY_TYPE_NAME','POPULATION','POPULATION_GROUP_ID','SUBURBAN_AREA_FLAG',
                         'COUNTY_NAME']]

offense_type = pd.read_csv('NIBRS_OFFENSE_TYPE.csv')
offense_type = offense_type[['OFFENSE_TYPE_ID','OFFENSE_NAME','CRIME_AGAINST','OFFENSE_CATEGORY_NAME']]
#offense_type.head()

```

```python
agency_all.columns
```

```python
# Remove columns not to being considered for ML
incident_df = incident_all.drop(['NIBRS_MONTH_ID','CARGO_THEFT_FLAG','SUBMISSION_DATE',
                             'REPORT_DATE_FLAG','INCIDENT_STATUS','DATA_HOME',
                             'CLEARED_EXCEPT_DATE','ORIG_FORMAT','DID'],axis=1)
incident_df.head()
```

```python
# Remove columns not to being considered for ML
# Rename some columns to clarify if demographic information is related to the arrestee or victim
arrestee_df = arrestee_all.drop(['DATA_YEAR','ARRESTEE_SEQ_NUM','AGE_ID','RESIDENT_CODE','UNDER_18_DISPOSITION_CODE',
                                 'CLEARANCE_IND','AGE_RANGE_LOW_NUM','AGE_RANGE_HIGH_NUM','OFFENSE_TYPE_ID'],axis=1)
arrestee_df = arrestee_df.rename(columns={'AGE_NUM': 'ARR_AGE_NUM', 'SEX_CODE': 'ARR_SEX_CODE','RACE_ID':'ARR_RACE_ID','ETHNICITY_ID':'ARR_ETHNICITY_ID'})
arrestee_df.head()
```

```python
# Remove columns not to being considered for ML
offense_df = offense_all.drop(['DATA_YEAR','ATTEMPT_COMPLETE_FLAG','NUM_PREMISES_ENTERED','METHOD_ENTRY_CODE'],axis=1)
offense_df.head()
```

```python
# Select only the offense type id's that account for the top 5 crime categories
offense_df = offense_df.loc[((offense_df['OFFENSE_TYPE_ID'] == 27) | (offense_df['OFFENSE_TYPE_ID'] == 44) | (offense_df['OFFENSE_TYPE_ID'] == 51)
                            | (offense_df['OFFENSE_TYPE_ID'] == 49)| (offense_df['OFFENSE_TYPE_ID'] == 5) | (offense_df['OFFENSE_TYPE_ID'] == 16)
                            | (offense_df['OFFENSE_TYPE_ID'] == 35) | (offense_df['OFFENSE_TYPE_ID'] == 58)| (offense_df['OFFENSE_TYPE_ID'] == 7)
                            | (offense_df['OFFENSE_TYPE_ID'] == 13)| (offense_df['OFFENSE_TYPE_ID'] == 14) | (offense_df['OFFENSE_TYPE_ID'] == 18)
                            | (offense_df['OFFENSE_TYPE_ID'] == 23)| (offense_df['OFFENSE_TYPE_ID'] == 45) | (offense_df['OFFENSE_TYPE_ID'] == 47))]
offense_df.head()
```

```python
offense_df.count()
```

```python
# Remove columns not to being considered for ML
offender_df = offender_all.drop(['DATA_YEAR','AGE_ID','AGE_RANGE_LOW_NUM','AGE_RANGE_HIGH_NUM','RACE_ID','ETHNICITY_ID','AGE_NUM','SEX_CODE'], axis=1)
offender_df.head()
```

```python
# Remove columns not to being considered for ML
# Rename some columns to clarify if demographic information is related to the arrestee or victim
victim_df = victim_all.drop(['DATA_YEAR','ASSIGNMENT_TYPE_ID', 'ACTIVITY_TYPE_ID','OUTSIDE_AGENCY_ID',
                     'AGE_ID','RESIDENT_STATUS_CODE','AGE_RANGE_LOW_NUM','AGE_RANGE_HIGH_NUM'],axis=1)

victim_df = victim_df.rename(columns={'AGE_NUM': 'VICT_AGE_NUM', 'SEX_CODE': 'VICT_SEX_CODE','RACE_ID':'VICT_RACE_ID','ETHNICITY_ID':'VICT_ETHNICITY_ID'})
victim_df.head()
```

```python
print(incident_df.count())
print(arrestee_df.count())
print(offense_df.count())
print(offender_df.count())
print(victim_df.count())
```

```python
#Remove records with NaN values
df_lst = [incident_df,arrestee_df,offense_df,offender_df,victim_df]
for i in df_lst:
    i.dropna(inplace=True)
```

```python
print(incident_df.count())
print(arrestee_df.count())
print(offense_df.count())
print(offender_df.count())
print(victim_df.count())
```

```python
# Consolidate all dataframes into a single dataframe using the merge function
merge_1 = pd.merge(offense_df,incident_df, on = "INCIDENT_ID")
merge_2 = pd.merge(merge_1,offender_df, on = "INCIDENT_ID")
merge_3 = pd.merge(merge_2, arrestee_df, on = "INCIDENT_ID")
merge_4 = pd.merge(merge_3, victim_df, on = "INCIDENT_ID")
merge_5 = pd.merge(merge_4, offense_type, on = "OFFENSE_TYPE_ID")
merge_6 = pd.merge(merge_5,agency_all, on="AGENCY_ID")
merge_6.head()
```

```python
cols = list(merge_6.columns.values)
print(cols)
```

```python
#Reorder the columns so it is easier to understand the data
df = merge_6[['OFFENSE_ID','INCIDENT_ID','OFFENDER_ID','ARRESTEE_ID','VICTIM_ID','AGENCY_ID','DATA_YEAR','INCIDENT_DATE',
             'ARREST_DATE','INCIDENT_HOUR','OFFENSE_TYPE_ID','OFFENSE_NAME','CRIME_AGAINST','OFFENSE_CATEGORY_NAME','LOCATION_ID','CLEARED_EXCEPT_ID','ARREST_TYPE_ID',
              'OFFENDER_SEQ_NUM','MULTIPLE_INDICATOR','ARR_AGE_NUM', 'ARR_SEX_CODE', 'ARR_RACE_ID', 
              'ARR_ETHNICITY_ID', 'VICTIM_SEQ_NUM', 'VICTIM_TYPE_ID', 'VICT_AGE_NUM', 'VICT_SEX_CODE', 
              'VICT_RACE_ID', 'VICT_ETHNICITY_ID','PUB_AGENCY_NAME', 'AGENCY_TYPE_NAME', 'POPULATION', 'POPULATION_GROUP_ID', 
              'SUBURBAN_AREA_FLAG', 'COUNTY_NAME']]
df.head()
```
