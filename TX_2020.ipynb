```python
import pandas as pd
```

```python
#Read in select csvs
incident = pd.read_csv("NIBRS_incident.csv")
offense = pd.read_csv("NIBRS_OFFENSE.csv")
agency = pd.read_csv("agencies.csv")
offense_type = pd.read_csv("NIBRS_OFFENSE_TYPE.csv")
```

```python
incident_df = incident.drop(['DATA_YEAR','NIBRS_MONTH_ID','CARGO_THEFT_FLAG','SUBMISSION_DATE',
                             'REPORT_DATE_FLAG','INCIDENT_STATUS','DATA_HOME',
                             'CLEARED_EXCEPT_DATE','ORIG_FORMAT','DID'],axis=1)
agency_df = agency[["AGENCY_ID","PUB_AGENCY_NAME","AGENCY_TYPE_NAME",
                    "POPULATION","SUBURBAN_AREA_FLAG","POPULATION_GROUP_CODE","COUNTY_NAME"]]
offense_df = offense.drop(['DATA_YEAR','ATTEMPT_COMPLETE_FLAG','NUM_PREMISES_ENTERED','METHOD_ENTRY_CODE','LOCATION_ID'],axis=1)
offense_type_df = offense_type.drop(['CT_FLAG','HC_FLAG','HC_CODE'],axis=1)
```

```python
agency_df.head()
```

```python
#Remove records with NaN values
df_lst = [incident_df,offense_df]
for i in df_lst:
    i.dropna(inplace=True)
```

```python
df_1 = pd.merge(incident_df,offense_df, on="INCIDENT_ID")
df_2 = pd.merge(df_1,agency_df,on="AGENCY_ID")
df_3 = pd.merge(df_2,offense_type_df,left_on="OFFENSE_TYPE_ID",right_on="OFFENSE_TYPE_ID")
df_3.head()
```

```python
df_3.to_csv("Texas_2020.csv",index=False)
```
